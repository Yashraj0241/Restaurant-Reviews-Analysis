### Step 12. Create a billing report (dataframe) which has columns as 'Order_Id', 'Total_Bill', 'Total Bill After Discount'. Restaurant owner has decided to give discount of 5% on orders of above $50 total cost.

```python
df
```

```python
# df["item_price"]=df["item_price"].apply(lambda x:float(x.strip("$")))
# df
```

```python
billing_report=df.groupby("order_id")["item_price"].sum().reset_index()
billing_report
```

```python
billing_report["Total_Bill_After_Discount"]=billing_report["item_price"].apply(lambda x: x*0.95 if x>50 else x)
billing_report.rename(columns={"item_price":"Total_bill"},inplace=True)
billing_report
```
